```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
base_path = "./pcv_data/"
file_name = "empire.jpg"

im = np.array(Image.open(base_path + file_name))
print(im.shape)
plt.figure(figsize=(8,6))
plt.imshow(im)

x = [100, 100, 400, 400]
y = [200, 500, 200, 500]

plt.plot(x, y,'r*')
plt.plot(x[:2],y[:2])
```

```python
plt.figure(figsize=(8,6))
plt.imshow(im)

x = [100, 100, 400, 400]
y = [200, 500, 200, 500]

plt.plot(x, y,'r*')
plt.plot(x[:2],y[:2])
plt.axis("off")
```

```python
# Image contours and histgrams

im = np.array(Image.open(base_path + file_name).convert('L'))
plt.figure()
plt.contour(im, origin="image")
plt.axis("equal")
```

```python
plt.hist(im.flatten(),bins=128)
plt.show()
```
